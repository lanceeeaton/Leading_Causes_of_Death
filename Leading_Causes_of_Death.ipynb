```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/lanceeeaton/Leading_Causes_of_Death/master/NCHS_-_Leading_Causes_of_Death__United_States.csv')
df = df.drop(['113 Cause Name'], axis = 1)

us_totals = df.loc[df['State'] == 'United States'] # US total by year state and cause

all_causes = df.loc[df['Cause Name'] == 'All causes'] # total by year and state
all_causes = all_causes.loc[df['State'] != 'United States']


df = df.loc[np.logical_and(df['State'] != 'United States',df['Cause Name'] != 'All causes')]

causes = df['Cause Name'].unique()
states = df['State'].unique()
years = df['Year'].unique()
```

```
grid = sns.FacetGrid(all_causes,hue='State',size = 10)
sns.plt.title('Death over Time by State')
grid.map(plt.scatter,'Year','Deaths')
grid.add_legend()
```

```
grid = sns.FacetGrid(all_causes,hue='State',size = 10)
sns.plt.title('Adj-Death over Time by State')
grid.map(plt.scatter,'Year','Age-adjusted Death Rate')
grid.add_legend()
```
