```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

death_state_year_cause = pd.read_csv('https://raw.githubusercontent.com/lanceeeaton/Leading_Causes_of_Death/master/death_state_year_cause.csv')

causes = death_state_year_cause['Cause Name'].unique()
states = death_state_year_cause['State'].unique()
years = death_state_year_cause['Year'].unique()
```

```
grid = sns.FacetGrid(death_state_year_cause,hue='State',size = 10)
sns.plt.title('Death over Time by State')
grid.map(plt.scatter,'Year','Deaths')
grid.add_legend()
```

```
grid = sns.FacetGrid(death_state_year_cause,hue='State',size = 10)
sns.plt.title('Adj-Death over Time by State')
grid.map(plt.scatter,'Year','Age-adjusted Death Rate')
grid.add_legend()
```
